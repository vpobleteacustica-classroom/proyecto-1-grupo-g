```python
# 1. Mapeo de Piezas a Nota MIDI 
# Usamos números MIDI (0-127). 60 = Do central.
mapa_piezas_nota = {
    chess.PAWN: 60,   # Do4
    chess.KNIGHT: 62, # Re4
    chess.BISHOP: 64, # Mi4
    chess.ROOK: 65,   # Fa4
    chess.QUEEN: 67,  # Sol4
    chess.KING: 69,   # La4
}

# 2. Parámetros de Síntesis (Eventos)
DURACION_NOTA = 240      # Duración en 'ticks' MIDI (tempo-dependiente)
VELOCITY_NORMAL = 80     # Intensidad (volumen) normal (0-127)
VELOCITY_CAPTURA = 110   # Intensidad para capturas
VELOCITY_JAQUE = 127     # Intensidad máxima para jaques
DURACION_ARPEGIO = 120   # Duración más corta para el arpegio del enroque
VELOCITY_ESPECIAL = 127  # Velocity máxima para eventos especiales

print("Mapeo de sonidos definido.")

# 3. Mapeo de Columna (a-h) a Paneo MIDI (0-127)
mapa_columna_pan = {
    0: 0,   # Columna 'a'
    1: 18,  # Columna 'b'
    2: 36,  # Columna 'c'
    3: 54,  # Columna 'd'
    4: 72,  # Columna 'e'
    5: 90,  # Columna 'f'
    6: 108, # Columna 'g'
    7: 127  # Columna 'h'
}

# 2. Mapeo de Fila (1-8) a Octava (Offset de Nota)
mapa_fila_octava = {
    0: 0,   # Fila '1'
    1: 2,   # Fila '2'
    2: 4,   # Fila '3'
    3: 6,   # Fila '4'
    4: 8,   # Fila '5'
    5: 10,  # Fila '6'
    6: 12,  # Fila '7' 
    7: 14   # Fila '8'
}

print("Mapeo de Filas y Columnas definido.")
```

```python
# 1. Configurar el archivo MIDI
mid = mido.MidiFile()
track = mido.MidiTrack()
mid.tracks.append(track)

# 2. Añadir Metadatos (buena práctica)
track.append(mido.MetaMessage('set_tempo', tempo=mido.bpm2tempo(140))) # 140 BPM
track.append(mido.MetaMessage('track_name', name='Sonificacion Ajedrez ACUS220'))

# 3. Iterar sobre la partida y generar eventos 
board = game.board()
print("Iniciando sonificación...")

for move in game.mainline_moves():

    # --- Coordenadas del movimiento ---
    columna_destino = chess.square_file(move.to_square) # `chess.square_file(sq)` nos da la columna (0-7)
    fila_destino = chess.square_rank(move.to_square)    # `chess.square_rank(sq)` nos da la fila (0-7)
    
    # --- Calcular Paneo y Octava ---
    pan_valor = mapa_columna_pan.get(columna_destino, 64) 
    octava_offset = mapa_fila_octava.get(fila_destino, 0)
    track.append(mido.Message('control_change', channel=0, control=10, value=pan_valor, time=0))

    # --- Detección de Promoción ---
    if move.promotion:
        print(f"Promoción detectada: {move}")
        # Obtenemos la nota de la nueva pieza
        pieza_promovida = move.promotion
        nota_base = mapa_piezas_nota.get(pieza_promovida, 67) 
        nota_final = nota_base + octava_offset

        # Tocamos un "Acorde Mayor"
        track.append(mido.Message('note_on', note=nota_base, velocity=VELOCITY_ESPECIAL, time=0))
        track.append(mido.Message('note_on', note=nota_base + 4, velocity=VELOCITY_ESPECIAL, time=0))
        track.append(mido.Message('note_on', note=nota_base + 7, velocity=VELOCITY_ESPECIAL, time=0))
        
        track.append(mido.Message('note_off', note=nota_base, velocity=0, time=DURACION_NOTA * 2))
        track.append(mido.Message('note_off', note=nota_base + 4, velocity=0, time=0))
        track.append(mido.Message('note_off', note=nota_base + 7, velocity=0, time=0))
        
        board.push(move) # Aplicamos el movimiento
        continue 

    # --- Detección de Enroque ---
    if board.is_castling(move):
        print(f"Enroque detectado: {move}")
        nota_rey = mapa_piezas_nota.get(chess.KING) + octava_offset
        nota_torre = mapa_piezas_nota.get(chess.ROOK) + octava_offset
        
        # Arpegio 
        track.append(mido.Message('note_on', note=nota_rey, velocity=VELOCITY_ESPECIAL, time=0))
        track.append(mido.Message('note_off', note=nota_rey, velocity=0, time=DURACION_ARPEGIO))
        track.append(mido.Message('note_on', note=nota_torre, velocity=VELOCITY_ESPECIAL, time=0))
        track.append(mido.Message('note_off', note=nota_torre, velocity=0, time=DURACION_ARPEGIO))

        board.push(move) # Aplicamos el movimiento
        continue 

    # --- Identificar la pieza y el evento ---
    pieza_movida = board.piece_at(move.from_square).piece_type
    
    # --- Definir parámetros acústicos base ---
    nota_base = mapa_piezas_nota.get(pieza_movida, 60) # Default a Peón
    velocity = VELOCITY_NORMAL
    duracion = DURACION_NOTA

    # --- Aplicar la nueva espacialidad ---
    nota_final = nota_base + octava_offset

    # --- Lógica de Captura ---
    if board.is_capture(move):
        velocity = VELOCITY_CAPTURA
    board.push(move)
    
    # --- Lógica de Jaque ---
    if board.is_check():
        velocity = VELOCITY_JAQUE
        duracion = int(DURACION_NOTA / 2) # Staccato

    # --- Añadir el evento MIDI ---
    track.append(mido.Message('note_on', note=nota_final, velocity=velocity, time=0))
    track.append(mido.Message('note_off', note=nota_final, velocity=0, time=duracion))

# 4. Guardar el archivo MIDI 

output_filename = 'partida_sonificada.mid'
mid.save(output_filename)

print("¡Proceso completado!")
print(f"Archivo MIDI guardado en: {output_filename}")
```
